```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline

x1 = np.linspace(0,10,25)

# create a plot figure
fig = plt.figure()

plt.plot(x1,np.sin(x1),'-')
plt.plot(x1,np.cos(x1),'--')
#plt.plot(x1,np.tan(x1),'*')
plt.show()
```

```python
plt.subplot(3,1,1)  #rows,columns,panel number
plt.plot(x1,np.sin(x1),'--')

plt.subplot(2,1,2)
plt.plot(x1,np.cos(x1))
plt.show()
```
